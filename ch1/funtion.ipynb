### 함수

- 자주 사용하는 한 부분
- 입력 => 출력

```
def 함수명(메개변수1, 매개변수2.....):
    해야할 일들
    해야할 일들
    return 실행 후 결과값,....
    
```


````python
# +

a = 3
b = 4
a + b
````

````python
# 함수명은 add 이고, 입력으로 a, b 2개의 갑을 받으며
# 리턴(출력)값은 2개의 입력을 더한 값이다

def add(a, b):
    return a + b
````

````python
print(add(3,4))
print(add(8,9))
print(add(15,30))
````

````python
# 입력값이 없는 함수

def say():
    return "HI"
````

````python
say()
````

````python
#리턴이 없는 함수

def add2(a, b):
    print(f"{a} + {b} = {a + b}")
````

````python
add2(5,7)
add2(12,7)
add2(19,27)
````

````python
# 입력값도 없고 리턴값도 없는 함수

def say2():
    print("HI")

say2()
````

````python
# 빼기 함수
# sub()

def sub(a, b):
    return a - b

print(sub(7,3))

#매개변수 지정 후 호출
print(sub(a=8,b=3))
print(sub(b=8,a=13))
````

````python
# add
# add(2,3)
# add(2,3,4,5)
# add(2,3,4,5,6,7)

# 가변 매개변수
# 입력값이 몇 개가 될 지 모르는 경우
# *


#def 함수명(*변수명):
#   해야할일

def add(*args):
    result = 0
    for i in args:
        result = result + i
    print(result) 
````

````python
add(2,3)
add(2,3,4,5)
add(2,3,4,5,6,7)

````
