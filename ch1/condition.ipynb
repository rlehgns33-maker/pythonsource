'숫자가 짝수면 짝수라고 출력, 아니면 홀수라고 출력'

```python
num = 5

if num % 2 == 0:
    print("짝수")
else:
    print("홀수")
```

```python
# SyntaxError
# if num = 2:
#       print("짝수")


#  >, < , >=, <= ,== !=  : True, False
# in, not in
a = 200
# a가 100보다 작으면 a는 100보다 작다 출력

if a < 100:
    print("a는 100배보다 작다")
else:
    print("a는 100보다 크다")

# a가 100보다 크거나 200 이하라면 a는 100과 200 사이의 수이다 출력

if 100 < a and a <= 200:
    print("a는 100과 200사이의 수이다")
```

```python
# 점수가 90 이상이면 A, 80 이상이면 B, 70 이상이면 C, 나머지는 F

score = 99

if score >= 90:
    print("A")
elif score >=80:
    print("B")
elif score >=70:
    print("C")
else:
    print("F")
```

```python
a = 12
b = 6
c = 19

a,b,c = 12,6,19

#가장 큰 숫자 출력

if a > b and a > c:
    print(a)
elif b>a and b>c:
    print(b)
else:
    print(c)
```

```python
max = a
if max < b:
    max = b

if max < c:
    max = c

print("가장 큰 수", max)
```

```python
#가장 작은 수
min = a


if min > b:
    min = b

if min > c:
    min = c

print("가장 작은 수", min)
```
