```python
# try ~ except ~ else

try:
    age = int(input("나이 입력"))
except:
    print("입력이 정확하지 않습니다.")
else: # 오류가 없다면 
    if age <= 18:
        print("미성년자는 출입금지입니다.")
    else:
        print("환영합니다")
```

```python
# 예외 클래스 생성 가능
class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 별명입니다."
    


def say_nick(nick):
    if nick =="바보":
        raise MyError()
    print(nick)


try:
    say_nick("천사")
    say_nick("바보")
except MyError as e:
    print(e)
```

```python
# 문제 1. 랜덤 비밀번호 생성기

# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
# choice()
chars = "ABCDEFGHIJKMNOPQRSTUVWXYZabcdefg~~~~~0123456789!@#$%"
# 비밀너호 길이는 4 이상만 허용
# 숫자가 아닌 값을 입력하면 예외 처리한다

from random import random, choice, sample, shuffle, randrange,randint, uniform
import random

password =[]

try:
    number = int(input("생성할 비밀번호의 길이"))
    i = 0
    while i < number:
        password = password + random.choice(chars)
        i += i
    print(password)
except:
    print("숫자만 입력하시오.")
else: # 오류가 없다면 
    if number < 4:
        print("비밀번호가 4글자 미만입니다")
    else:
        print("비밀번호가 생성되었습니다")
```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 2001-08-25
# input() => '2001-08-25' => 날짜 형식
import datetime

now = datetime.datetime.now()

try:
    birth = int(input("생년월일을 입력하시오 ex)2001-08-25"))
except:
    print("입력이 정확하지 않습니다.")
else: # 오류가 없다면 
    if (now - birth) < 0:
        print("날짜를 잘못 입력했습니다")
    else:
        print(f"당신의 나이는 {now - birth}입니다")


print(now - birth)
```
