```python
# 예외 클래스 생성 가능
class MyError(Exception):
    def __str__(self):
        return "허용되지 않는 별명입니다."
    


def say_nick(nick):
    if nick =="바보":
        raise MyError()
    print(nick)


try:
    say_nick("천사")
    say_nick("바보")
except MyError as e:
    print(e)
```

```python
# 문제 1. 랜덤 비밀번호 생성기

# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
# choice()
chars = "ABCDEFGHIJKMNOPQRSTUVWXYZabcdefg~~~~~0123456789!@#$%"
# 비밀너호 길이는 4 이상만 허용
# 숫자가 아닌 값을 입력하면 예외 처리한다

from random import random, choice, sample, shuffle, randrange,randint, uniform
import random

password=[]

try:
    number = int(input("생성할 비밀번호의 길이"))
    i = 0
    while i < number:
        word = random.choice(chars)
        password.append(word) 
        i += 1
except:
    print("숫자만 입력하시오.")
else: # 오류가 없다면 
    if number < 4:
        print("비밀번호가 4글자 미만입니다")
    else:
        print("비밀번호가 생성되었습니다")
        #print(f"{password}")
        # 리스트 => 문자열로 변경
        print(f"{"".join(password)}")
```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 2001-08-25
# input() => '2001-08-25' => 날짜 형식
import datetime


# 오늘날짜
now = datetime.datetime.now()

try:
    # 생년월일 입력
    birth_str = input("생년월일을 입력하시오 ex)2001-08-25")
    # '2001-08-25'

    birth = datetime.datetime.strptime(birth_str,"%Y-%m-%d")

    # 2026 - 2001
    # now 년도만 추출
    age = now.year() - birth.year()
    if (now.month, now.day) < (birth.month, birth.day):
        age -= 1
    print(f"당신의 나이는 {age}입니다")

except:
    print("날짜 형식이 올바르지 않습니다")





```

```python
# 다음과 같은 csv 파일이 있다.
# 이름,국어,영어,수학
# 홍길동,90,80,70
# 김철수,85,70,80
# 이영희,92,88,76
# 성춘향,93,82,77

# csv 파일을 읽어 학생별 평균 점수 출력
# 홍길동 : 80.0
# csv.DictReader() 사용, 평균은 둘째자리까지
import csv

list1 = [
    {"이름":"홍길동","국어":90,"영어":80,"수학":70},
    {"이름":"김철수","국어":85,"영어":70,"수학":80},
    {"이름":"이영희","국어":92,"영어":88,"수학":76},
    {"이름":"성춘향","국어":93,"영어":82,"수학":77}

]

with open("./data/student.csv","w",encoding="utf-8",newline="")as f:
    headers = ['이름','국어','영어','수학']
    writer = csv.DictWriter(f, fieldnames=headers)
    # 해더작성
    writer.writeheader()
    #writer.writerow({"이름":"홍길동","국어":90,"영어":80,"수학":70})
    #writer.writerow({"이름":"김철수","국어":85,"영어":70,"수학":80})
    #writer.writerow({"이름":"이영희","국어":92,"영어":88,"수학":76})
    #writer.writerow({"이름":"성춘향","국어":93,"영어":82,"수학":77})
    for row in list1:
        writer.writerow(row)



with open("./data/student.csv","r",encoding="utf-8",newline="")as f:
    reader = csv.DictReader(f)
    #print(reader)

    for row in reader:
        #print(row)
        avg = (int(row.get('국어')) + int(row.get('영어')) + int(row.get('수학'))) / 3

        print(f"{row.get('이름')} : {avg:.2f}")
    
   
```
