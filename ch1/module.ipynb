```python
# random : 난수(규칙이 없는 임의의 수)를 발생시키는 모듈

import random

# random() : 0.0 ~ 1.0 사이의 실수 증에서 난수
# 0.0 <= random.random() < 1.0
print(random.random())
# 시작 숫자 <= random.random() < 끝 숫자
print(random.uniform(10,20))
# uniform(최소,최대) : 지정한 두 실수 사이의 난수

# #randrange(stop)
#randrange(start,stop)
#randrange(start,stop,step)
print(random.randrange(10))
#choice(시퀀스) : 리스트,튜플,문자열
print(random.choice([1,2,3,4,5,6,7]))
# shuffle()
list1 = list(range(1,8))
random.shuffle(list1)
print(list1)


# sample(시퀀스.개수) : 리스트,튜플,문자열
print(random.sample(list1, k=2))
print(random.sample("python", k=2))
```

```python
from random import random, choice, sample, shuffle, randrange,randint, uniform

# random : 난수(규칙이 없는 임의의 수)를 발생시키는 모듈

import random

# random() : 0.0 ~ 1.0 사이의 실수 증에서 난수
# 0.0 <= random.random() < 1.0
print(random.random())
# 시작 숫자 <= random.random() < 끝 숫자
print(uniform(10,20))
# uniform(최소,최대) : 지정한 두 실수 사이의 난수

# #randrange(stop)
#randrange(start,stop)
#randrange(start,stop,step)
print(randrange(10))
#choice(시퀀스) : 리스트,튜플,문자열
print(choice([1,2,3,4,5,6,7]))
# shuffle()
list1 = list(range(1,8))
shuffle(list1)
print(list1)


# sample(시퀀스.개수) : 리스트,튜플,문자열
print(sample(list1, k=2))
print(sample("python", k=2))
```

```python
# 쉼표로 구분된 파일
# 혹은 다양한 기호로 분리
import csv

with open("./data/sample1.csv","r") as f:
    reader = csv.DictReader(f)
    #print(reader)
    next(reader)
    for c in reader:
        for k, v in c.items():
            print(k, v)
```

```python
with open("./data/sample3.csv","w",encoding="utf-8",newline="")as f:
    headers = ['이름','국어','영어']
    writer = csv.DictWriter(f, fieldnames=headers)
    # 해더작성
    writer.writeheader()
    writer.writerow({"이름":"홍길동","국어":88,"영어":92})
    writer.writerow({"이름":"성춘향","국어":87,"영어":90})
```

```python
import json

with open("./data/myinfo1.json","r",encoding="utf-8",)as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
dict1 = {'name': '성춘향', 'birth': '0927', 'age': 24}

with open("./data/myinfo1.json","w",encoding="utf-8",)as f:
    json.dump(dict1, f,ensure_ascii=False)
```

```python
# loads() : json 문자열 => 파이썬 자료형(dict)으로 변환
# dumps() : 파이썬 자료형(dict) => json 문자열로 변환


str1 = """{
    "name": "홍길동",
    "birth": "0727",
    "age": 24
}
"""


data = json.loads(str1)
print(data)
print(type(data))


print(data['name'])
print(data.get('name'))
```

```python
dict1 = """{
    "name": "홍길동",
    "birth": "0727",
    "age": 24
}
"""

data = json.dumps(dict1)
print(data)
print(type(data))

```
