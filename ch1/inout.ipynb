```python
# 사직연산 함수
# 결과 리턴



def calculator(num1, num2, op):
    if op == "+":
        return num1 + num2
    elif op == "-":
        return num1 - num2
    elif op == "*":
        return num1 * num2
    elif op == "/":
        return num1 / num2
    elif op == "//":
        return num1 // num2
    

# 숫자 2개 입력 받기, 연산자 기호 (+,-,*,...)입력받기
num1 = int(input("첫번째 숫자 입력"))
num2 = int(input("두번째 숫자 입력"))

op = input("기호 입력")

print(calculator(num1, num2, op))

```

```python
values = input("두 숫자와 연사잔 기호 입력 ( ex 15 10 +)")

# 문자(공백을 기준으로 분리)
# '15 10 +'

#result = values.split()
#print(result)


num1,num2,op = values.split()
print(calculator(int(num1), int(num2), op))
```

```python
# 15,10,+

values = input("두 숫자와 연산자 기호 입력 ( ex 15,10,+)")

num1,num2,op = values.split(",")
print(calculator(int(num1), int(num2), op))
```
