###  file 읽고 쓰기

|파일 열기 모드 | 설명 |
|------------|------|
|'r'|**읽기(Read)모드** : 파일의 내용을 읽기만 할때 사용. 파일이 존재하지 않는 경우 오류 발생.|
|'w'|**쓰기(Write)모드** : 파일의 내용을 새로 쓸때 사용. 파일이 존재하는 경우 내용 사젝 후 새로 작성. 없는 경우 새로 생성.|
|'a'|**추가(Append)모드** : 기준 파일의 마지막에 내용을 추가할 때 사용, 파일이 없는 경우 새로 생성.|

```python
# score.txt 내용을 읽은 후 평균 구한 후 출력

scores = []
with open("./data/score.txt","r",encoding="utf-8") as f:
    for num in f:
        scores.append(int(num))

    print(f"{sum(scores)/len(scores)}.")
```

```python
# sample.txt 파일을 읽어서 총합과 평균을 구한 후 총합과 평균을 result.txt 파일 작성


sample = []
with open("./data/sample.txt","r",encoding="utf-8") as f:
    for num in f:
        sample.append(int(num))

    print(f"총합 : {sum(sample)}.")
    print(f"평균 : {sum(sample)/len(scores)}.")


with open("./data/result.txt","w",encoding="utf-8") as f:
    #f.write(f"총합 : {sum(sample)}.\n")
    #f.write(f"평균 : {sum(sample)/len(scores)}.\n")
    f.write("총합 : %d\n" % sum(sample))
    f.write("평균 : %.2f" % (sum(sample)/len(scores)))
```

```python
# fruits.txt 읽어서 단어의 개수 출력
# apple 3
# banana 2

# apple banana orange
# 단어 분리 : split() => ['apple','banana','orange']

# dict 추가
dict1 = {}

with open("./data/fruits.txt","r", encoding="utf-8") as f:
    for line in f:
        #print(line)
        words = line.split()
        #print(words)
        for word in words:
            # print(word)

            if word in dict1:
                dict1[word] += 1 
            else:
                dict1[word] = 1 

for fruit, count in dict1.items():
    print(f"{fruit} : {count}")
```

```python
# 1000 명의 이름, 키, 몸무게 info.txt 읽은 후
# BMI 지수 계산
# BMI = 체중 / 키의 제곱
# BMI가 18.5 미만일 때 저체중
# BMI가 18.5 ~ 22.9일 때 정상체중
# BMI가 23 ~ 24.9일 때 과체중
# BMI가 25이상일 때 비만


# 출력문
# 이름 : 가타
# 몸무개 : 
# 키 :
# BMI : 
# 결과 : 

name,weight,height = "가타, 44, 186".split(", ")


with open ("./data/info.txt","r",encoding="utf-8") as f:
    for line in f:
        name,weight,height = line.split(", ")
        #print(name,weight,height)
        weight = int(weight)
        height = int(height)

        BMI = (int(weight)/(int(height)/100)**2)
        if BMI < 18.5:
            result = "저체중"
        elif BMI <= 22.9:
            result = "정상체중"
        elif BMI <= 24.9:
            result = "과체중"
        else:
            result = "비만"



        print("\n".join(["이름 : {}", "몸무게 : {}", "키 : {}", "BMI : {:.1f}", "결과 : {}"]).format(name,weight,height,BMI,result))
        print()

        
```

```python
# origin.txt 읽어서 암호화 시킨 후 encr.txt 저장
# 1. 암호화 선택 : origin.txt 읽어서 변수에 담기 => 그걸 chr(ord("안")+100) 저장
# 2. 복호화 선택 : 암호화된 파일을 읽은 후 원래대로 출력


while True:
    no = int(input("1.암호화 | 2.복호화 | 3.종료 중 원하는 메뉴의 번호를 입력하세요"))

    if no == 1:
        with open ("./data/origin.txt", "r",encoding="utf-8") as f:
            contents = f.read()
    

        with open ("./data/encr.txt", "w",encoding="utf-8") as f:
            for c in contents:
                f.write(chr(ord(c)+100))
    elif no == 2:
        with open ("./data/encr.txt", "r",encoding="utf-8") as f:
            contents = f.read()
            for c in range(len(contents)):
                print(chr(ord(contents[c]) - 100), end="")
    else:
        break


```
