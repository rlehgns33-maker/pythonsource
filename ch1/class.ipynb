```python
# 자동차 클래스 설계
# 속성 : 색상, 속도
# 동작:  전진, 후진, 속도 올리기, 속도 내리기 .... => 멤버 메소드

class Car:
    def __init__(self,color,speed):
        # 인스턴스(멤버) 변수
        self.color = color
        self.speed = speed

    # 인스턴스(멤버) 메서드(함수)
    def upSpeed(self, value):
        self.speed += value


    def downSpeed(self, value):
        self.speed -= value


    #0x000002060CEB5950 => 보고싶은 문자로 변경
    def __str__(self):
        return f"Car(color={self.color}, speed={self.speed})"
```

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner), 잔액(balance)
#   메소드 : deposit(money) : 잔액 = 잔액 + 입금액
#   메소드 : withraw(money) : 잔액 = 잔액 - 출금액
#   메소드 : showBalance() : 현재 잔액 출력


class Account:

    def __init__(self,owner,balance = None):
        self.owner =owner
        self.balance =balance
        

    def deposit(self, money):
        self.balance += money

    def withraw(self, money):
        if self.balance < money:
            print("잔액 부족")
        else:
            self.balance -= money


    def showBalance(self):
        print(f"현재 잔액 출력 : {self.balance}")


owner1 = Account("홍길동", 100000)
owner1.showBalance()

owner1.withraw(2000)
owner1.showBalance()

owner1.deposit(3000)
owner1.showBalance()
```

```python
# Book 클래스 작성
# 클래스 변수 : book_count
# 멤버 변수 : 제목(title), 저자(author)
#    메소드 : showInfo()=> 제목 : 파이썬
#                         저자 : 홍길동
# 현재 책 권수 출력

class Book:

    book_count = 0

    def __init__(self,title,author):
        self.title = title
        self.author = author
        Book.book_count += 1

    def showInfo(self):
        print(f"제목:{self.title} \n저자:{self.author}")


book1 = Book("파이썬","홍길동")
book2 = Book("자바","성춘향")

book1.showInfo()
book2.showInfo()
print(f"현재 책 권수 출력 {Book.book_count}권")
```

```python
# Order 클래스 작성
# 클래스 변수 : total_order
# 멤버 변수 : 메뉴(menu), 가격(price), 수량(count)
# 메소드 : getTotalPrice() = 가격 x 수량 : 총 주문금액 반환
#         showOrder() : 출력문 작성
#         메뉴 : 아메리카노
#         가격 : 4000원
#         수량 : 2개
#         총 주문 금액 : 8000원

class Order:
    total_order = 0

    def __init__(self,menu,price,count):
        self.menu = menu
        self.price = price
        self.count = count
        Order.total_order += 1


    def getTotalPrice(self):
        return self.price * self.count
    
    def showOrder(self):
        print(f"메뉴 : {self.menu}\n가격 : {self.price}\n수량 : {self.count}\n 총 주문 금액 : {self.getTotalPrice}")


order1 = Order("아메리카노",4500,2)
order2 = Order("카페라뗴",4800,1)

order1.showOrder()
print()
order2.showOrder()
              

print(f"총 주문 금액 : {Order.total_order}")
    
```

- 코드 재사용성, 확장

```python
class Parent:
    def __init__(self):
        self.value = "테스트"
        print("Parent 클래스의 __init__() 호출")

    def test(self):
        print("Parent 클래스의 test() 호출")

class Child(Parent):
    pass

child = Child()
child.test()

print(child.value)
```

```python
# super() : 상위 클래스

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child 클래스의 __init__ 호출")

child = Child()
child.test()

print(child.value)
```

```python
class Animal:
    def __init__(self,name):
        self.name = name

    def eat(self):
        print(f"{self.name}가 먹이를 먹는다")

    def sound(self):
        print(f"{self.name}가 소리를 낸다")


class Dog(Animal):
    def __init__(self, name, kind):
        super().__init__(name)
        self.kind = kind

    def fetch(self):
        print(f"{self.name}이 공을 물어옵니다.")

dog = Dog("초코","푸들")
dog.eat()
dog.sound()
dog.fetch()
```

```python
# Audio : 멤버변수 : power, volume / 메소드 : 전원 조절, 볼륨조절 + tune()
# TV : 멤버변수 : power, volume, size / 메소드 : 전원 조절, 볼륨조절 + watch()
class AudioVisual:
    def __init__(self,power,volume):
        self.power = power
        self.volume = volume
    
    def switch(self, on_off):
        self.power = on_off
    
    def setVolume(self, volume):
        self.volume = volume

class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        print("La la la...." if self.power else "turn it on")

class TV(AudioVisual):
    def __init__(self, power, volume,size):
        super().__init__(power, volume)
        self.size = size
    def watch(self):
        print("have fun!!" if self.power else "switch on")

tv = TV(False, 12, 50)
# tv 켜기
tv.switch(True)
# tv 보기
tv.watch()

audio = Audio(True, 15)
audio.setVolume(10)
audio.tune()
audio.switch(False)

```

#### 오버라이딩
- 부모 클래스에 있는 메서드를 자식 클래스에서 같은 이름으로 다시 정의 해서 동작 내용 변경

```python
class Animal:
    def __init__(self,name):
        self.name = name

    def sound(self):
        print(f"{self.name}가 소리를 낸다")

class Dog(Animal):
    def sound(self):
        print(f"{self.name} : 멍멍!")
        #return super().sound()
    
class Cat(Animal):
    def sound(self):
        # 오버라이딩 시 동일한 부모 메소드를 호출하고 싶다면
        super().sound()
        print(f"{self.name} : 야옹!")
    
    
dog = Dog("초코")
cat = Cat("콜라")

dog.sound()
cat.sound()
```

```python
# Employee 부모 클래스 :
# 멤버 변수 : 이름(name), 급여(pay)
#     메서드 : showInfo() => 이름, 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
# apthem : getTotalPay() : 급여 + 보너스 출력

class Employee:
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay

    def showInfo(self):
        print(f"이름 : {self.name}, 급여 : {self.pay}")


class Manager(Employee):
    def __init__(self, name, pay, bonus):
        super().__init__(name, pay)
        self.bonus = bonus

    def getTotalPay(self):
        print(f"급여 + 보너스 출력 : {self.pay + self.bonus}")


user1 = Employee("홍길동", 100000)
user1.showInfo()

user2 = Manager("홍길동", 100000, 33333)

user2.showInfo()
user2.getTotalPay()
```

```python
# shape 부모 클래스
# 메서드 : area() ==> "도형의 넓이" 출력

# Rectangle 자식 클래스
# 멤버 변수 : width, height
# 메서드 : area() => 직사각형의 넓이 출력

# Circle 자식 클래스
# 멤버 변수 : 반지름(radius)
#    메서드 : area() => 원의 넓이 출력


class Shape:
    def area(self):
        print("도형의 넓이")

class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__()
        self.width = width
        self.height = height

    def area(self):
        print(f"사각형 넓이 : {self.width * self.height}")


class Circle(Shape):
    def __init__(self,radius):
        super().__init__()
        self.radius= radius

    def area(self):
        print(f"원의 넓이 : {self.radius **2 * 3.14}")


r = Rectangle(12,10)
r.area()

c = Circle(10)
c.area()


```
